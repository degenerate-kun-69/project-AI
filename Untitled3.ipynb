```python
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
import matplotlib.pyplot as plt

# Load restaurant data
restaurants = pd.read_csv('restaurants_data.csv')
restaurants['Price for Two'] = restaurants['Price for Two'].str.replace('Rs', '').str.replace(',', '').astype(float)
restaurants['Rating'] = restaurants['Rating'].astype(float)

# Load user data
users = pd.read_csv('user_data.csv')
users['Usual Spending Budget'] = users['Usual Spending Budget'].str.replace('Rs', '').str.replace(',', '').astype(float)

# Load cuisine mapping data (book1.csv)
cuisine_mapping = pd.read_csv('book1.csv', header=None, names=['Unique Cuisines', 'Label', 'Code'])

# Merge user data with restaurant features for training
data = users.merge(restaurants, how='cross')

# Define feature columns and target column
feature_columns = ['Usual Spending Budget', 'Acceptable Rating', 'Rating', 'Price for Two']
target_column = 'target'

# Define the target: 1 if restaurant rating >= acceptable rating, else 0
data['target'] = (data['Rating'] >= data['Acceptable Rating']).astype(int)

# Extract features and target
X = data[feature_columns].values
y = data['target'].values

# Normalize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Convert to PyTorch tensors
X_tensor = torch.tensor(X_scaled, dtype=torch.float32)
y_tensor = torch.tensor(y, dtype=torch.float32).unsqueeze(1)

# Define a simple neural network model (from the first version)
class RecommenderNN(nn.Module):
    def __init__(self, input_dim):
        super(RecommenderNN, self).__init__()
        self.layers = nn.Sequential(
            nn.Linear(input_dim, 64),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(64, 32),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(32, 1),
            nn.Sigmoid()  # Output probabilities
        )

    def forward(self, x):
        return self.layers(x)

# Initialize the model, loss function, and optimizer
input_dim = X_tensor.shape[1]
model = RecommenderNN(input_dim)
criterion = nn.BCELoss()  # Binary Cross-Entropy Loss for classification
optimizer = optim.Adam(model.parameters(), lr=0.001)

# Training loop
epochs = 100
for epoch in range(epochs):
    model.train()
    optimizer.zero_grad()
    predictions = model(X_tensor)
    loss = criterion(predictions, y_tensor)
    loss.backward()
    optimizer.step()

    if (epoch + 1) % 10 == 0:  # Print loss every 10 epochs
        print(f'Epoch {epoch + 1}/{epochs}, Loss: {loss.item()}')

# Function to calculate classification metrics (from the second version)
def calculate_classification_metrics(model, X, y, threshold=0.5):
    """
    Calculate classification metrics for a trained PyTorch model.

    Args:
        model: The trained PyTorch model.
        X: Features (PyTorch tensor).
        y: True labels (PyTorch tensor).
        threshold: Threshold for binary classification (default: 0.5).

    Returns:
        A dictionary with accuracy, precision, recall, F1 score, and ROC-AUC score.
    """
    # Get predictions
    model.eval()
    with torch.no_grad():
        probabilities = model(X).numpy().flatten()  # Get predicted probabilities

    # Convert probabilities to binary predictions using the threshold
    predictions = (probabilities >= threshold).astype(int)

    # Convert true labels to NumPy array
    y_true = y.numpy().flatten()

    # Calculate metrics
    metrics = {
        "Accuracy": accuracy_score(y_true, predictions),
        "Precision": precision_score(y_true, predictions),
        "Recall": recall_score(y_true, predictions),
        "F1 Score": f1_score(y_true, predictions),
        "ROC-AUC": roc_auc_score(y_true, probabilities)  # Use probabilities for AUC
    }

    return metrics

# Calculate and display classification metrics
metrics = calculate_classification_metrics(model, X_tensor, y_tensor)
print("\nClassification Metrics:")
for metric, value in metrics.items():
    print(f"{metric}: {value:.4f}")

# Define weighted feature columns and update feature importance (from the third version)
feature_weights = {
    'Usual Spending Budget': 0.2,  # Budget weight
    'Acceptable Rating': 0.3,  # User's acceptable rating weight
    'Rating': 0.4,  # Restaurant rating weight
    'Price for Two': 0.1,  # Minor influence
}

def preprocess_features(data, feature_weights):
    """
    Adjust feature values based on weights.
    """
    # Scale budget to account for fuzzy matches (e.g., Gaussian scaling)
    data['Usual Spending Budget'] = np.exp(
        -((data['Usual Spending Budget'] - data['Price for Two'])**2) / (2 * (data['Price for Two'].std()**2))
    )

    # Normalize and apply weights
    weighted_features = {}
    for feature, weight in feature_weights.items():
        weighted_features[feature] = data[feature] * weight

    return pd.DataFrame(weighted_features)

# Function to predict and plot probabilities for a new user (from the third version)
def predict_and_plot_for_user(user_input, restaurants_df, cuisine_mapping_df):
    """
    Predicts probabilities for a new user and filters restaurants based on cuisine preferences
    with prioritization.
    """
    preferred_cuisines = user_input.get('Cuisines', None)
    if preferred_cuisines:
        preferred_cuisines = [c.strip().lower() for c in preferred_cuisines.split(',')]
        cuisine_codes = cuisine_mapping_df[
            (cuisine_mapping_df['Unique Cuisines'].str.lower().isin(preferred_cuisines)) |
            (cuisine_mapping_df['Label'].str.lower().isin(preferred_cuisines))
        ]['Code'].astype(str).tolist()
    else:
        cuisine_codes = []

    if cuisine_codes:
        restaurants_df['Cuisines'] = restaurants_df['Cuisines'].astype(str)
        filtered_restaurants = restaurants_df[
            restaurants_df['Cuisines'].apply(lambda x: any(code in x.split(', ') for code in cuisine_codes))
        ]
    else:
        filtered_restaurants = restaurants_df.copy()

    user_df = pd.DataFrame([user_input]).merge(filtered_restaurants, how='cross')

    user_weighted_features = preprocess_features(user_df, feature_weights)
    user_X = user_weighted_features.values
    user_X_scaled = scaler.transform(user_X)
    user_X_tensor = torch.tensor(user_X_scaled, dtype=torch.float32)

    model.eval()
    with torch.no_grad():
        probabilities = model(user_X_tensor).numpy().flatten()

    user_df['Probability'] = probabilities
    sorted_user_df = user_df.sort_values(by='Probability', ascending=False)

    sorted_user_df['Cuisines_y'] = sorted_user_df['Cuisines_y'].apply(
    lambda x: ', '.join(sorted(set(x.split(', '))))  # Remove duplicates and sort cuisines
)


    # Plotting and output
    plt.figure(figsize=(12, 6))
    plt.bar(sorted_user_df['Restaurant Name'], sorted_user_df['Probability'], color='skyblue')
    plt.xticks(rotation=90)
    plt.xlabel('Restaurant Name')
    plt.ylabel('Probability of Liking')
    plt.title('Probability of User Liking Each Restaurant')
    plt.tight_layout()
    plt.show()

    # Display top recommendations
    top_recommendations = sorted_user_df[['Restaurant Name', 'Cuisines_y', 'Probability']].head(5)
    print("Top Recommended Restaurants:")
    print(top_recommendations.to_string(index=False))

# Test new user input
new_user_input = {
    'Usual Spending Budget': 1500,
    'Acceptable Rating': 4.0,
    'Cuisines': 'Chinese, north indian'
}

predict_and_plot_for_user(new_user_input, restaurants, cuisine_mapping)
```

```python
import pandas as pd
import torch
import matplotlib.pyplot as plt

# Assuming you have loaded your book1.csv into a DataFrame (e.g., cuisine_mapping_df)
book1 = pd.read_csv('book1.csv')  # Load the CSV

# Create a dictionary to map cuisine codes to their respective names
cuisine_code_to_name = dict(zip(book1['Code'].astype(str), book1['Unique Cuisines']))

def predict_and_plot_for_user(user_input, restaurants_df, cuisine_mapping_df):
    """
    Predicts probabilities for a new user and filters restaurants based on cuisine preferences,
    prioritizing restaurants that serve all cuisines from the user input.
    """
    # Get preferred cuisines
    preferred_cuisines = user_input.get('Cuisines', None)
    if preferred_cuisines:
        preferred_cuisines = [c.strip().lower() for c in preferred_cuisines.split(',')]
        cuisine_codes = cuisine_mapping_df[
            (cuisine_mapping_df['Unique Cuisines'].str.lower().isin(preferred_cuisines)) |
            (cuisine_mapping_df['Label'].str.lower().isin(preferred_cuisines))
        ]['Code'].astype(str).tolist()
    else:
        cuisine_codes = []

    # Filter based on cuisine preferences
    if cuisine_codes:
        restaurants_df['Cuisines'] = restaurants_df['Cuisines'].astype(str)
        filtered_restaurants = restaurants_df[
            restaurants_df['Cuisines'].apply(lambda x: any(code in x.split(', ') for code in cuisine_codes))
        ]
    else:
        filtered_restaurants = restaurants_df.copy()

    # Merge with user input
    user_df = pd.DataFrame([user_input]).merge(filtered_restaurants, how='cross')

    # Preprocess features with weighted importance
    user_weighted_features = preprocess_features(user_df, feature_weights)
    user_X = user_weighted_features.values
    user_X_scaled = scaler.transform(user_X)
    user_X_tensor = torch.tensor(user_X_scaled, dtype=torch.float32)

    # Predict probabilities
    model.eval()
    with torch.no_grad():
        probabilities = model(user_X_tensor).numpy().flatten()

    user_df['Probability'] = probabilities
    sorted_user_df = user_df.sort_values(by='Probability', ascending=False)

    # Clean duplicate cuisine codes in Cuisines_y
    sorted_user_df['Cuisines_y'] = sorted_user_df['Cuisines_y'].apply(
        lambda x: ', '.join(sorted(set(x.split(', '))))  # Remove duplicates and sort cuisines
    )

    # Replace codes with cuisine names for the final output only
    sorted_user_df['Cuisines_y'] = sorted_user_df['Cuisines_y'].apply(
        lambda x: ', '.join([cuisine_code_to_name.get(code.strip(), code) for code in x.split(', ')])
    )

    # Calculate the number of cuisines from user input that each restaurant serves
    def count_matching_cuisines(restaurant_cuisines):
        return sum(1 for cuisine in preferred_cuisines if cuisine.lower() in restaurant_cuisines.lower())

    # Add a new column for the number of matching cuisines
    sorted_user_df['Matching Cuisines'] = sorted_user_df['Cuisines_y'].apply(count_matching_cuisines)

    # Adjust probability based on the number of matching cuisines
    def adjust_probability(row):
        matched_cuisines = row['Matching Cuisines']
        return row['Probability'] * (matched_cuisines / len(preferred_cuisines))

    sorted_user_df['Adjusted Probability'] = sorted_user_df.apply(adjust_probability, axis=1)

    # Sort first by matching cuisines (descending) and then by adjusted probability (descending)
    sorted_user_df = sorted_user_df.sort_values(by=['Matching Cuisines', 'Adjusted Probability'], ascending=[False, False])

    # Plotting and output
    plt.figure(figsize=(15, 10))
    plt.bar(sorted_user_df['Restaurant Name'], sorted_user_df['Adjusted Probability'], color='skyblue')
    plt.xticks(rotation=90)
    plt.xlabel('Restaurant Name')
    plt.ylabel('Adjusted Probability of Liking')
    plt.title('Adjusted Probability of User Liking Each Restaurant')
    plt.tight_layout()
    plt.show()

    # Display top recommendations with cuisine names
    top_recommendations = sorted_user_df[['Restaurant Name', 'Cuisines_y', 'Adjusted Probability']].head(5)
    print("Top Recommended Restaurants:")
    print(top_recommendations.to_string(index=False))

# Example usage
new_user_input = {
    'Usual Spending Budget': 4000,
    'Acceptable Rating': 3.7,
    'Cuisines': 'Chinese'
}

predict_and_plot_for_user(new_user_input, restaurants, cuisine_mapping)
```
